##### Step 1: Install and Import Required Libraries

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
df_comp = df[['ConvertedCompYearly']]

df_comp['ConvertedCompYearly'] = pd.to_numeric(
    df_comp['ConvertedCompYearly'],
    errors='coerce'
)

df_comp = df_comp.dropna() # drop nulls

# create histogram
plt.figure(figsize=(10, 6))
df['ConvertedCompYearly'].plot.hist(bins=2000, edgecolor='black', alpha=0.7) # 'bins' controls the number of bars
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)')
plt.xlabel('Yearly Compensation')
plt.ylabel('Count')
plt.xlim(0, 600000)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
df_ft = df[df['Employment'].str.contains('Employed, full-time', na=False)]

median_full_time_comp = df_ft['ConvertedCompYearly'].median()
print('Median compensation for full-time employees:', median_full_time_comp)
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
df_country = df.dropna(subset=['Country', 'ConvertedCompYearly'])

# exclude countries with minimal data
min_n = 200
valid_countries = df_country['Country'].value_counts()
valid_countries = valid_countries[valid_countries >= min_n].index
df_country = df_country[df_country['Country'].isin(valid_countries)].copy()

# rename 'United Kingdom of Great Britain and Northern Ireland'
df_country['Country'] = df_country['Country'].replace({
    'United Kingdom of Great Britain and Northern Ireland':
        'United Kingdom',
    'United States of America':
         'USA'   
})

# create boxplot
countries = df_country['Country'].unique()
df_country = [df_country.loc[df_country['Country'] == c, 'ConvertedCompYearly'] for c in countries]

plt.figure(figsize=(12, 12))
plt.boxplot(df_country, tick_labels=countries, showfliers=True) # remove outliers
plt.yscale('log')  # makes the plot readable
plt.xticks(rotation=45, ha='right')
plt.ylabel('Yearly Compensation (log scale)')
plt.title('Compensation Distribution by Country')
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
q1 = df['ConvertedCompYearly'].quantile(0.25) # bottom of interquartile range
q3 = df['ConvertedCompYearly'].quantile(0.75) # top of interquartile range

#calculate interquartile range
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr # lower limit
upper_bound = q3 + 1.5 * iqr # upper limit

df_cleaned = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)] #create dataframe excluding without outliers

print("Original dataset size:", df.shape)
print("Dataset size without outliers:", df_cleaned.shape)
print("Outliers removed:", df.shape[0] - df_cleaned.shape[0])
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
# check for missing values
print('Missing ConvertedCompYearly:' ,df['ConvertedCompYearly'].isna().sum())
print('Missing WorkExp:' ,df['WorkExp'].isna().sum())
print('Missing JobSatPoints_1:' ,df['JobSatPoints_1'].isna().sum())

# impute all the empty rows in the column Work Exp & JobSatPoints_1  with the mode
workexp_mode = df['WorkExp'].mode()[0]
df['WorkExp'] = df['WorkExp'].fillna(workexp_mode)

jobsat_mode = df['JobSatPoints_1'].mode()[0]
df['JobSatPoints_1'] = df['JobSatPoints_1'].fillna(jobsat_mode)

# create matrix of correlations
df_subset = df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']]
matrix = df_subset.corr()

# create heatmap
plt.figure(figsize=(8, 6)) # Set the figure size for better readability
sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Heatmap of ConvertedCompYearly, WorkExp, and JobSatPoints_1 Correlations')
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
# Scatter plot of ConvertedCompYearly and WorkExp
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, x='WorkExp', y='ConvertedCompYearly')
plt.title('Compensation Against Work Experience')
plt.xlabel('Work Experience')
plt.ylabel('Converted Compensation Yearly')
plt.show()

# Scatter plot of ConvertedCompYearly and JobSatPoints_1
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, x='JobSatPoints_1', y='ConvertedCompYearly')
plt.title('Compensation Against Job Satisfaction')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Converted Compensation Yearly')
plt.show()
```
