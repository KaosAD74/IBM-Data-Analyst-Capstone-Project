### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
duplicates = df.duplicated(subset=df.columns.difference(["ResponseId"])).sum() # exclude the unique identifier
print(duplicates)
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
subset_cols = df.columns.difference(["ResponseId"]) # exlude unique identifier

num_duplicates_before = df.duplicated(subset=subset_cols).sum() # count duplicates before cleaning

df_dedupe = df.drop_duplicates(subset=subset_cols, keep="first") #drop duplicates

num_duplicates_after = df_dedupe.duplicated(subset=subset_cols).sum() # count duplicates after cleaning

print ("Before: ", num_duplicates_before)
print ("After: ", num_duplicates_after)
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
print(df.isnull().sum().sort_values(ascending=False))  # count null values
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
remotework_missing = df['RemoteWork'].isnull().sum()
print("Count of RemoteWork Missing Values:", {remotework_missing})
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
remotework_value_counts = df.groupby(['RemoteWork']).size()
print("Count of Remote Work Values:", remotework_value_counts)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
most_freq_value = remotework_value_counts.idxmax()
print(most_freq_value)
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
import numpy as np

df['RemoteWork'] = df['RemoteWork'].replace(np.nan, most_freq_value)
remotework_value_counts2 = df.groupby(['RemoteWork']).size() # check for changed number in hybrid
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
!pip install -U seaborn

import seaborn as sns
import matplotlib.pyplot as plt

df_salary = df.dropna(subset=['ConvertedCompYearly']) # drop missing rows
df_salary['ConvertedCompYearly'].describe() # get descriptive statistics

#create histogram with log scale
plt.figure(figsize=(10, 6))
sns.histplot(df_salary['ConvertedCompYearly'], bins=50, kde=True, log_scale=True)
plt.title('Distribution of Converted Compensation (Log Scale)')
plt.xlabel('Converted Compensation (log scale)')
plt.ylabel('Count')
plt.show()
```
